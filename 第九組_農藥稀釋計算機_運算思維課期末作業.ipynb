## 農藥稀釋計算機

此筆記本將實現以下三個功能：
1.  **單位換算標準化**: 將農地面積單位（公畝、分、公頃）自動換算為標準單位（例如：平方公尺），並將農藥用量單位（公升、毫升、公斤、公克）標準化。
2.  **自動安全提醒**: 若農藥量大於安全用量標準，則釋出安全提醒。
3.  **面積與容量規劃**: 依據個人使用的噴霧器容量自動分配單次配藥量，或根據特定面積計算總施藥成本與預算。

### 功能一: 單位換算標準化

此功能將農地面積單位統一換算為**平方公尺 (m²)**，並為農藥用量提供標準化轉換（例如：公升與毫升、公斤與公克之間的轉換）。

**面積單位換算比例:**
*   1 公畝 (a) = 100 平方公尺
*   1 分 = 969.917 平方公尺
*   1 公頃 (ha) = 10,000 平方公尺

**農藥用量單位換算比例:**
*   1 公升 (L) = 1000 毫升 (mL)
*   1 公斤 (kg) = 1000 公克 (g)

```python
def convert_area_to_sq_meters(value, unit):
    """
    將農地面積換算為平方公尺。

    Args:
        value (float): 面積數值。
        unit (str): 面積單位 ('公畝', '分', '公頃', '平方公尺'，不區分大小寫)。

    Returns:
        float: 換算後的平方公尺值。
    Raises:
        ValueError: 如果單位不支援。
    """
    unit_lower = unit.lower()
    if unit_lower == '公畝':
        return value * 100
    elif unit_lower == '分':
        return value * 969.917
    elif unit_lower == '公頃':
        return value * 10000
    elif unit_lower == '平方公尺' or unit_lower == 'm²' or unit_lower == 'sqm':
        return value
    else:
        raise ValueError(f"不支援的面積單位: {unit}. 支援的單位有 '公畝', '分', '公頃', '平方公尺'.")

def convert_pesticide_volume(value, unit_from, unit_to):
    """
    將農藥體積或質量單位進行換算。

    Args:
        value (float): 農藥數值。
        unit_from (str): 原始單位 ('公升', '毫升', '公斤', '公克'，不區分大小寫)。
        unit_to (str): 目標單位 ('公升', '毫升', '公斤', '公克'，不區分大小寫)。

    Returns:
        float: 換算後的目標單位值。
    Raises:
        ValueError: 如果單位不支援或單位類型不匹配。
    """
    unit_from_lower = unit_from.lower()
    unit_to_lower = unit_to.lower()

    volume_units = {'公升', 'l', '毫升', 'ml'}
    mass_units = {'公斤', 'kg', '公克', 'g'}

    is_volume_from = unit_from_lower in volume_units
    is_mass_from = unit_from_lower in mass_units
    is_volume_to = unit_to_lower in volume_units
    is_mass_to = unit_to_lower in mass_units

    if (is_volume_from and is_mass_to) or (is_mass_from and is_volume_to):
        raise ValueError("無法在體積和質量單位之間直接轉換，除非提供密度資訊。")

    if is_volume_from and is_volume_to:
        if unit_from_lower == '公升' or unit_from_lower == 'l':
            if unit_to_lower == '毫升' or unit_to_lower == 'ml':
                return value * 1000
            elif unit_to_lower == '公升' or unit_to_lower == 'l':
                return value
        elif unit_from_lower == '毫升' or unit_from_lower == 'ml':
            if unit_to_lower == '公升' or unit_to_lower == 'l':
                return value / 1000
            elif unit_to_lower == '毫升' or unit_to_lower == 'ml':
                return value
    elif is_mass_from and is_mass_to:
        if unit_from_lower == '公斤' or unit_from_lower == 'kg':
            if unit_to_lower == '公克' or unit_to_lower == 'g':
                return value * 1000
            elif unit_to_lower == '公斤' or unit_to_lower == 'kg':
                return value
        elif unit_from_lower == '公克' or unit_from_lower == 'g':
            if unit_to_lower == '公斤' or unit_to_lower == 'kg':
                return value / 1000
            elif unit_to_lower == '公克' or unit_to_lower == 'g':
                return value

    raise ValueError(f"不支援的農藥單位轉換: 從 {unit_from} 到 {unit_to}")

# 測試單位換算功能
print("--- 單位換算測試 ---")
area_in_sqm = convert_area_to_sq_meters(1, '公頃')
print(f"1 公頃 = {area_in_sqm} 平方公尺")
area_in_sqm = convert_area_to_sq_meters(5, '分')
print(f"5 分 = {area_in_sqm} 平方公尺")

volume_in_ml = convert_pesticide_volume(2, '公升', '毫升')
print(f"2 公升 = {volume_in_ml} 毫升")
weight_in_kg = convert_pesticide_volume(500, '公克', '公斤')
print(f"500 公克 = {weight_in_kg} 公斤")
```

### 功能二: 自動安全提醒

此功能將根據預設的安全用量標準，判斷農藥的使用量是否過高。如果計算出的農藥用量超過安全標準，系統將發出警告。

```python
def check_safety_alert(pesticide_amount, safe_limit, unit):
    """
    檢查農藥用量是否超過安全標準，並發出提醒。

    Args:
        pesticide_amount (float): 實際使用的農藥量。
        safe_limit (float): 安全用量上限。
        unit (str): 農藥單位 (例如: '毫升/公升', '克/公頃' 等，用於提示訊息)。

    Returns:
        bool: 如果超過安全標準則為 True，否則為 False。
    """
    if pesticide_amount > safe_limit:
        print(f"⚠️ 安全提醒: 農藥用量 ({pesticide_amount} {unit}) 超過安全標準 ({safe_limit} {unit})！請立即檢查。")
        return True
    else:
        print(f"✅ 農藥用量 ({pesticide_amount} {unit}) 在安全範圍內 ({safe_limit} {unit})。")
        return False

# 測試安全提醒功能
print("\n--- 安全提醒測試 ---")
# 假設每公頃安全上限為 500 毫升
check_safety_alert(450, 500, '毫升/公頃') # 安全範圍內
check_safety_alert(600, 500, '毫升/公頃') # 超過安全標準
```

### 功能三: 面積與容量規劃

此功能協助使用者根據噴霧器容量或特定施藥面積來規劃農藥用量。它包括：

1.  **單次配藥量計算**: 根據噴霧器容量和推薦的稀釋倍數，計算每次加藥所需的農藥量。
2.  **總施藥成本與預算**: 根據總施藥面積、農藥單價和稀釋倍數，估算所需的農藥總量及總成本。

```python
def calculate_single_dose(sprayer_capacity, dilution_ratio):
    """
    根據噴霧器容量和稀釋倍數，計算單次配藥所需的農藥量。

    Args:
        sprayer_capacity (float): 噴霧器容量 (例如: 公升)。
        dilution_ratio (float): 稀釋倍數 (例如: 1000 代表 1:1000 稀釋)。

    Returns:
        float: 單次配藥所需的農藥量 (單位與噴霧器容量單位一致)。
    """
    if dilution_ratio <= 0:
        raise ValueError("稀釋倍數必須大於 0。")
    return sprayer_capacity / dilution_ratio

def calculate_total_cost_and_budget(total_area_sqm, pesticide_per_area_unit, pesticide_unit_price, area_unit_for_pesticide='平方公尺'):
    """
    根據總施藥面積和農藥資訊，計算所需的農藥總量及總成本。

    Args:
        total_area_sqm (float): 總施藥面積 (平方公尺)。
        pesticide_per_area_unit (float): 每單位面積所需的農藥量。
        pesticide_unit_price (float): 農藥每單位價格 (例如: 1000 元/公升)。
        area_unit_for_pesticide (str): pesticide_per_area_unit 对应的面积单位 (例如: '平方公尺', '公頃')。

    Returns:
        tuple: (total_pesticide_needed, total_cost)
               total_pesticide_needed (float): 所需農藥總量。
               total_cost (float): 總成本。
    """
    # 先將 pesticide_per_area_unit 轉換為每平方公尺的用量
    if area_unit_for_pesticide.lower() == '公頃':
        pesticide_per_sqm = pesticide_per_area_unit / 10000 # 從每公頃轉為每平方公尺
    elif area_unit_for_pesticide.lower() == '平方公尺':
        pesticide_per_sqm = pesticide_per_area_unit
    else:
        raise ValueError(f"不支援的農藥用量面積單位: {area_unit_for_pesticide}. 支援 '公頃' 或 '平方公尺'.")

    total_pesticide_needed = total_area_sqm * pesticide_per_sqm
    total_cost = total_pesticide_needed * pesticide_unit_price
    return total_pesticide_needed, total_cost

# 測試面積與容量規劃功能
print("\n--- 面積與容量規劃測試 ---")
# 噴霧器容量為 16 公升，稀釋倍數 1000 倍
dose = calculate_single_dose(16, 1000)
print(f"噴霧器容量 16 公升，稀釋 1:1000，單次配藥需農藥: {dose} 公升 ({dose*1000} 毫升)")

# 總施藥面積 1 公頃，每公頃需農藥 2 公升，每公升農藥 800 元
total_area = convert_area_to_sq_meters(1, '公頃') # 轉換為平方公尺
pesticide_per_ha = 2 # 每公頃 2 公升
pesticide_price_per_liter = 800 # 每公升 800 元

total_pesticide, total_cost = calculate_total_cost_and_budget(
    total_area_sqm=total_area,
    pesticide_per_area_unit=pesticide_per_ha,
    pesticide_unit_price=pesticide_price_per_liter,
    area_unit_for_pesticide='公頃'
)
print(f"總施藥面積 1 公頃 ({total_area} 平方公尺)，總計需農藥: {total_pesticide} 公升，總成本: {total_cost} 元")
```

### 功能四: 稀釋倍數計算

此功能將根據農藥原液用量與稀釋後的總水量，計算出實際的稀釋倍數。這有助於驗證配藥是否符合預期濃度。

```python
def calculate_dilution_ratio(pesticide_amount, total_diluted_volume, pesticide_unit='毫升', volume_unit='公升'):
    """
    計算農藥稀釋倍數。

    Args:
        pesticide_amount (float): 農藥原液用量。
        total_diluted_volume (float): 稀釋後的總體積（水 + 農藥）。
        pesticide_unit (str): 農藥用量的單位（例如: '毫升', '公升'）。
        volume_unit (str): 稀釋後總體積的單位（例如: '毫升', '公升'）。

    Returns:
        float: 稀釋倍數 (例如: 1000 代表 1:1000 稀釋)。
    Raises:
        ValueError: 如果輸入值無效或單位轉換失敗。
    """
    # 將所有單位轉換為毫升進行計算
    pesticide_amount_ml = convert_pesticide_volume(pesticide_amount, pesticide_unit, '毫升')
    total_diluted_volume_ml = convert_pesticide_volume(total_diluted_volume, volume_unit, '毫升')

    if pesticide_amount_ml <= 0 or total_diluted_volume_ml <= 0:
        raise ValueError("農藥用量和稀釋後總體積都必須大於 0。")

    # 稀釋倍數 = 稀釋後的總體積 / 農藥原液用量
    # 例如: 1000 mL 水 + 1 mL 農藥 -> 總體積 1001 mL，稀釋倍數約 1000
    # 更精確地說，稀釋倍數通常指的是「藥液佔總體積的比例的倒數」
    # 或更常見的是「總水量 / 農藥量」
    # 這裡假設為 (總水量 / 農藥量)，因為噴霧器容量通常指的是總液體量，而非純水。
    # 如果 'total_diluted_volume' 已經包含了農藥量，則稀釋倍數是總體積與農藥量的比例。
    # 如果 'total_diluted_volume' 是純水，則應為 (total_diluted_volume + pesticide_amount_ml) / pesticide_amount_ml

    # 為了簡化，我們假設稀釋倍數為 '總稀釋液體量 / 農藥原液量'
    # 如果是水:藥液比例 N:1, 稀釋倍數就是 N+1
    # 如果是總液量:藥液比例 N:1, 稀釋倍數就是 N

    # 我們將採用 `總稀釋液體量 / 農藥原液量` (總體積與農藥原液的比值) 來表示稀釋倍數
    # 例如 16公升噴霧器，0.016公升農藥，稀釋倍數是 16 / 0.016 = 1000
    return total_diluted_volume_ml / pesticide_amount_ml

# 測試稀釋倍數計算功能
print("\n--- 稀釋倍數計算測試 ---")
# 假設在 16 公升的總稀釋液中，使用了 16 毫升的農藥
dilution = calculate_dilution_ratio(16, 16, pesticide_unit='毫升', volume_unit='公升')
print(f"使用 16 毫升農藥於 16 公升總稀釋液中，稀釋倍數為 1:{int(dilution)}")

# 另一個例子：500 毫升農藥於 50 公升總稀釋液中
dilution_2 = calculate_dilution_ratio(500, 50, pesticide_unit='毫升', volume_unit='公升')
print(f"使用 500 毫升農藥於 50 公升總稀釋液中，稀釋倍數為 1:{int(dilution_2)}")
```

### 功能五: 禁忌作物警告

此功能提供農藥使用時的禁忌作物警示。它依賴於一個預定義的農藥-作物禁忌列表。當使用者輸入農藥名稱和目標作物時，系統會檢查是否存在禁忌情況並發出警告。

**注意事項**：本範例中的禁忌列表僅供示範，實際應用應使用官方或專業的農藥安全數據。

```python
# 建立農藥與禁忌作物的資料庫 (僅供示範)
# 鍵為農藥名稱 (大寫)，值為禁忌作物列表 (大寫)
PESTICIDE_CONTRAINDICATIONS = {
    '農藥A': ['番茄', '草莓'],
    '農藥B': ['黃瓜', '甜椒'],
    '除草劑C': ['所有農作物'], # 除草劑通常會影響所有作物
    '殺菌劑X': ['葡萄', '蘋果']
}

def check_crop_contraindication(pesticide_name, target_crop):
    """
    檢查特定農藥對目標作物是否有禁忌。

    Args:
        pesticide_name (str): 農藥名稱。
        target_crop (str): 目標作物名稱。

    Returns:
        bool: 如果是禁忌作物則為 True，否則為 False。
    """
    pesticide_name_upper = pesticide_name.upper()
    target_crop_upper = target_crop.upper()

    if pesticide_name_upper in PESTICIDE_CONTRAINDICATIONS:
        forbidden_crops = PESTICIDE_CONTRAINDICATIONS[pesticide_name_upper]
        if '所有農作物' in forbidden_crops or target_crop_upper in forbidden_crops:
            print(f"⚠️ 禁忌作物警告: 農藥 '{pesticide_name}' 不建議用於作物 '{target_crop}'！請謹慎使用或諮詢專家。")
            return True
    print(f"✅ 農藥 '{pesticide_name}' 目前沒有針對作物 '{target_crop}' 的明確禁忌警告。")
    return False

# 測試禁忌作物警告功能
print("\n--- 禁忌作物警告測試 ---")
check_crop_contraindication('農藥A', '番茄')
check_crop_contraindication('農藥B', '玉米')
check_crop_contraindication('除草劑C', '小麥')
check_crop_contraindication('殺菌劑X', '梨子')
```
